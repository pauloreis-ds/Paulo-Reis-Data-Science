### 1.1 Adding some features
        Using some texts to make the graph more readable.
        Pay attention to the sintax of the methods.

```python
x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values)

plt.title('Adding title')   # New
plt.xlabel('X Horizontal Label')  # New
plt.ylabel('Y Vertical Label')   # New

plt.show()
```

        It is possible to manipulate even more those characteristics using other parameters

```python
x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values)

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   # New
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24})   # New
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   # New

plt.show()
```

        Changing the X and Y ticks. In the chart above, the Y ticks, for exemple,
        scales up like 10, 15, 20, 25, 30, 35... what if we want to make it
        10, 20 ,30...

```python
x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values)

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4])  # New
plt.yticks([10,20,30,40])  # New

plt.show()
```

        Legend is very helpful when dealing with various objects in the same graph
        

```python
''' This is not a very good way to do it, depends on the situation '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values)

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend('red line legend')  # New

plt.show()
```

```python
''' but we rather use label in the plt.plot() '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values, label='2X line function') # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

```python
''' still in plt.plot(parameters) we can pass a color ('name' or ''#code') '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]

plt.plot(x_axis_values, y_axis_values, label='2X line function', color='blue') # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

```python
''' and change the marker '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]
#                                                                    '#color_code'
plt.plot(x_axis_values, y_axis_values, label='2X line function', color='#0000FF', marker='.') # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

        Marker parameter can also be manipulated

```python
''' and change the marker '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]
#                                                                    '#purple_code'
plt.plot(x_axis_values, y_axis_values, label='2X line function', color='#800080', marker='.',
        markersize=18, markeredgecolor="red") # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

```python
''' And finally, linestyle '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]
                                                                   
plt.plot(x_axis_values, y_axis_values, label='2X line function', color='#800080', marker='.',
        markersize=18, markeredgecolor="red", linestyle='--') # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

```python
''' Shorthand notation '''

x_axis_values = [1,2,3,4]
y_axis_values = [10,20,30,40]


# Shorthand notation:
# fmt = '[color][marker][line]'                                                                    
plt.plot(x_axis_values, y_axis_values, 'r^--', label='2X line function' ) # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

#### Adding another line
        It's pretty simple, in fact, we are just creating another
        chart but in the same space of the first chart

```python
x_axis_values = [1,2,3,4] 
y_axis_values = [10,20,30,40] 

x_line2 = [1, 1.5, 2, 2.5, 3, 3.5, 4]    # New
y_line2 = [1, 2.25, 4, 6.25, 9, 12.25, 16]  # New
                                                               
plt.plot(x_axis_values, y_axis_values, 'r^--', label='2X line function' ) 

plt.plot(x_line2, y_line2, 'b',label='second line') # New

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

plt.show()
```

```python
# And using plt.figure() we can change the figure size
plt.figure(figsize=(8,3))  # New

x_axis_values = [1,2,3,4] 
y_axis_values = [10,20,30,40] 

x_line2 = [1, 1.5, 2, 2.5, 3, 3.5, 4]    
y_line2 = [1, 2.25, 4, 6.25, 9, 12.25, 16]  
                                                               
plt.plot(x_axis_values, y_axis_values, 'r^--', label='2X line function' ) 

plt.plot(x_line2, y_line2, 'b',label='second line')

plt.title('Adding title', fontdict={'fontname':'Gulim', 'fontsize':30})   
plt.xlabel('X Horizontal Label', fontdict={'fontname':'SimSun', 'fontsize':24}) 
plt.ylabel('Y Vertical Label', fontdict={'fontname':'SimSun', 'fontsize':24})   

plt.xticks([1,2,3,4]) 
plt.yticks([10,20,30,40])  

plt.legend() 

# Creating a png file
plt.savefig('savefig.png')

plt.show()
```

## 2 - Bar chart
        The process is similar, but since it's a different kind of
        graph, it has some different features.

```python
categories = ['Class A', 'Class B', 'Class C']
categories_values = [10, 20, 30]

plt.figure(figsize=(6,5))  

plt.bar(categories,categories_values)

plt.show()
```

        We can add pattern to each class. But to do that, we need to use a little
        bit of object oriented approach

```python
categories = ['Class A', 'Class B', 'Class C']
categories_values = [10, 20, 30]

plt.figure(figsize=(6,5))  

# Creating an instance for the bar object. and now we can use some specific methods.
bars = plt.bar(categories,categories_values)

bars[0].set_hatch('/') # New

plt.show()
```

```python
''' And do the same thing with the other classes '''

categories = ['Class A', 'Class B', 'Class C']
categories_values = [10, 20, 30]

plt.figure(figsize=(6,5))  

bars = plt.bar(categories,categories_values)

bars[0].set_hatch('/') 
bars[1].set_hatch('O') # New
bars[2].set_hatch('*') # New


plt.show()
```

```python
plt.style.use('fivethirtyeight') # Just changing the style

categories = ['Class A', 'Class B', 'Class C']
categories_values = [10, 20, 30]

plt.figure(figsize=(6,5))  

bars = plt.bar(categories,categories_values)

bars[0].set_hatch('/') 
bars[1].set_hatch('O') 
bars[2].set_hatch('*') 

plt.savefig('my fig.png')

plt.show()
```
