##  1 - Basic line graph (gas prices)

```python
# plt.plot(x_axis_values, y_axis_values) in this case, x and y will be the columns values

plt.plot(gas.Year, gas.USA)

plt.show()
```

```python
# We can also plot two 'functions' in the same graph

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)


plt.show()
```

### 1.1 Readability
        Let's make our graph better to understand
##### Text information

```python
plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)
plt.plot(gas.Year, gas['South Korea']) # New

plt.title('TITLE HERE')       # New
plt.xlabel('X - HORIZONTAL  LABEL') # New
plt.ylabel('Y - VERTICAL  LABEL') # New


plt.show()
```

##### Legend
        To use use legends We need to label the plots

```python
plt.plot(gas.Year, gas.USA, label='USA')  # New
plt.plot(gas.Year, gas.Canada, label='Canada')   # New
plt.plot(gas.Year, gas['South Korea'], label='South Korea')  # New

plt.title('Gas Prices Over Time')       
plt.xlabel('Years') 
plt.ylabel('Prices') 

plt.legend()  # New

plt.show()
```

#### Beautifying

```python
# linestyle (shorthand notation)
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')  # New
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')  # New  
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')  # New
plt.plot(gas.Year, gas['Australia'], 'y.-', label='South Korea')  # New


plt.title('Gas Prices Over Time')       
plt.xlabel('Years')
plt.ylabel('Prices') 

plt.xticks(gas.Year[::3])  

plt.legend()  

plt.show()
```

```python
# fontstyle

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')  
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')   
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')  
plt.plot(gas.Year, gas['Australia'], 'y.-', label='South Korea') 


plt.title('Gas Prices Over Time', fontdict={'fontsize':24, 'fontname':'Impact'})  # New
plt.xlabel('Years', fontdict={'fontsize':18, 'fontname':'fantasy'})   # New
plt.ylabel('Prices', fontdict={'fontsize':18, 'fontname':'fantasy'})   # New

plt.xticks(gas.Year[::3])  

plt.legend()  

plt.show()
```

```python
# Finally, figure size and saving a png file.

plt.figure(figsize=(10,5)) # New

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')  
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')   
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')  
plt.plot(gas.Year, gas['Australia'], 'y.-', label='South Korea') 


plt.title('Gas Prices Over Time', fontdict={'fontsize':24, 'fontname':'Impact'})  # New
plt.xlabel('Years', fontdict={'fontsize':18, 'fontname':'fantasy'})   # New
plt.ylabel('Prices', fontdict={'fontsize':18, 'fontname':'fantasy'})   # New

plt.xticks(gas.Year[::3])  

plt.legend()  

plt.savefig('gas_prices.png') # New

plt.show()
```

## 2 - Bar chart (Fifa data)
        First of all, load the data into memory

```python
fifa = pd.read_csv("fifa_data.csv", )
fifa.head()
```

### Histograms

```python
plt.hist(fifa.Overall)

plt.show()
```

```python
''' Better visualization of the bins '''

bins = [40,50,60,70,80,90,100]  # New

plt.hist(fifa.Overall, bins=bins)  # New

plt.show()
```

```python
''' Adding text information '''

bins = [40,50,60,70,80,90,100]  

plt.hist(fifa.Overall, bins=bins)  

plt.title('Number of players')  # New
plt.xlabel('Skill Level')  # New
plt.ylabel('Distribution of Player skills in FIFA 2019')  # New

plt.show()
```

### Pie chart

```python
fifa['Preferred Foot'].head()
```

```python
# Getting only the count of Left foot
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

# Getting only the count of Left foot
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right])

plt.show()
```

##### Adding labels, text information and colors 

```python
# Labels and title

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']  # New
                                                     
plt.pie([left, right], labels = labels)

plt.title('Foot Preference of FIFA Players') # New

plt.show()
```

```python
# Percentage

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']  

plt.pie([left, right], labels = labels, autopct='%.2f %%') # New

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
# Colors

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']  
colors = ['#abcdef', '#aabbcc']  # New

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%') # New

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### 3 - Boxplot (comparing teams)

```python
# Getting the Overall skills of these team players
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

barcelona[0:6] # Is the first one Messi?
```

```python
fifa.head(2) # Yes, it is hehe
```

```python
plt.boxplot([barcelona, madrid])

plt.show()
```

#### Completing the chart

```python
# Adding a new team to compare, and labels

labels=['FC Barcelona','Real Madrid','NE Revolution']

plt.boxplot([barcelona, madrid,revs], labels=labels) # New

plt.show()
```

```python
''' As we can see, the NE Revolution is lower skilled '''

# Text Information

labels=['FC Barcelona','Real Madrid','NE Revolution']

plt.boxplot([barcelona, madrid,revs], labels=labels) 

plt.title('Professional Soccer Team Comparison') # New
plt.ylabel('FIFA Overall Rating') # New

plt.show()
```

```python
# Figure and font size

plt.figure(figsize=(7,9))

labels=['FC Barcelona','Real Madrid','NE Revolution']

plt.boxplot([barcelona, madrid,revs], labels=labels) 

plt.title('Professional Soccer Team Comparison', fontsize=18) # New
plt.ylabel('FIFA Overall Rating',fontsize=16) # New

plt.show()
```

```python
# Prettifying
plt.figure(figsize=(7,9))

labels=['FC Barcelona','Real Madrid','NE Revolution']

#                                                   (it enables filling with color)           (median line)
plt.boxplot([barcelona, madrid,revs], labels=labels,      patch_artist=True,        medianprops={'linewidth': 8})

plt.title('Professional Soccer Team Comparison', fontsize=18) 
plt.ylabel('FIFA Overall Rating',fontsize=16) 

plt.show()
```
