```python
# I'll use 's data from Wikipedia but you can choose any source you prefer (You just need to have one big unique text)
import wikipedia

# WorldCloud is our main object, STOPWORDS will treat "meaningless" words like 'I, and, with...'
from wordcloud import WordCloud, STOPWORDS

# Loading the image
import numpy as np
from PIL import Image

# Plotting
import matplotlib.pyplot as plt
```

```python
''' You don't really need to worry about this, because It really can be any other resource source  '''
def get_wiki(query):
    # Get best matching title for given query
    title = wikipedia.search(query)[0]

    # Get wikipedia page for selected title
    page = wikipedia.page(title)
    return page.content

python_article = get_wiki('Python programming language')

# It's a big article here's part of it
print(python_article[0:1000])
```

## WordClouds 

```python
# Address of your image (mine is in the same directory so its name and format is enough)
py_brand_mask = np.array(Image.open("py4.png"))
   
# Creating the wordloud
wordcloud = WordCloud(stopwords=STOPWORDS, # stopword already has a built-in set of stopwords 
                      background_color="black",
                      width=1000, height=1000, # of the wordloud
                      max_words=2000, # wordloud will use
                      mask=py_brand_mask, # Defining the image that will be used
                      max_font_size=200, min_font_size=1).generate(python_article)
#                     Font size will decrease from 200 until it is 1

# Plotting the wordcloud
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
```

### Let's try a White background

```python
py_brand_mask = np.array(Image.open("py4.png"))
   
wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color="white", # New
                      width=1000, height=1000, max_words=2000,
                      mask=py_brand_mask, max_font_size=200,
                      min_font_size=1).generate(python_article)

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
```
